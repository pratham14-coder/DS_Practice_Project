##  About the Dataset Columns

## Here’s a breakdown of the columns in the dataset:

1. buying: Buying price of the car (values: vhigh, high, med, low)

2. maint: Maintenance cost (values: vhigh, high, med, low)

3. doors: Number of doors (values: 2, 3, 4, 5more)

4. persons: Capacity in terms of persons (values: 2, 4, more)

5. lug_boot: Size of luggage boot (values: small, med, big)

6. safety: Safety rating (values: low, med, high)

7. class: Car acceptability class (target variable, values: unacc, acc, good, vgood)

#  Dataset Overview: Initial Insights
1.  Total Entries: 1728 rows

2. Total Features: 7 columns (including the target)

3. Data Types: All columns are of type object — meaning they are categorical

4. Missing Values:  None — All columns have 1728 non-null values

```python
df.describe()
```

## Insight
1.  __The target column class is highly imbalanced.__
*  unacc appears 1210 times, which is ~70% of the data.
*  Other classes like acc, good, vgood are underrepresented, which may affect model learning.
*  Insight: Use techniques like SMOTE, class weights, or stratified sampling to balance during training.
  
2.  __Buying & Maintenance Cost Bias__
   
* Both buying and maint have vhigh as the most common value, each appearing 432 times (25% of rows).
* Insight: Dataset might be skewed towards expensive cars, affecting generalization for low-budget ones.

3.  __Low Practicality Cars Dominate__
* doors = 2 and persons = 2 are the most frequent — both indicating small-sized cars.
* Insight: This dataset might over-represent compact or budget cars.

4. __Smaller Cars with Low Safety__
* lug_boot = small and safety = low are the most common, each in 576 entries.
* Insight: Could reflect a dataset focused on entry-level or low-tier vehicles, often unsuitable for families.

# Detecting the Outlier using the visulization

```python

plt.figure(figsize=(15, 15))  # Set the figure size for a 3x3 grid
plot_num = 1

# Loop through the selected columns
columns = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
for i in columns:
    plt.subplot(3, 3, plot_num)  # Create subplot in a 3x3 grid
    sns.boxplot(data=df, x=i)  # Bar plot for categorical data
    plt.title(i, fontsize=18)
    plt.xlabel(i, fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plot_num += 1

plt.tight_layout()  # Adjust layout for better spacing
plt.show() 
```

### Although outliers are detected in the class column, we do not replace or remove them since this column is the target variable. Modifying target values can negatively impact model training and prediction accuracy.

## EDA (Exploratory Data Analysis)

## __Univariant Analysis__

```python

# Set style for plots
sns.set(style="whitegrid")
plt.figure(figsize=(20, 15))

# List of categorical columns
cat_cols = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']

# Plot count plots in a 3x3 grid
for idx, col in enumerate(cat_cols):
    plt.subplot(3, 3, idx + 1)
    sns.countplot(data=df, x=col, hue=col, palette="Set2", legend=False)
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

## __Bivariant Analysis__

```python
# Categorical vs Categorical: Countplot for 'class' vs 'buying'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='buying', hue='class', palette="Set2")
plt.title('Class vs Buying')
plt.tight_layout()
plt.show()

# Categorical vs Categorical: Countplot for 'class' vs 'maint'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='maint', hue='class', palette="Set2")
plt.title('Class vs Maintenance')
plt.tight_layout()
plt.show()
```

# Defining X and Y

```python
X = df.drop(columns=["class"])
Y = df["class"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
X_train
```

```python
X_test
```

# Applying the data Imbalancing

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE()
```

```python
X_train_smote, Y_train_smote = sm.fit_resample(X_train, Y_train)
```

```python
from collections import Counter
```

```python
print("Before applying Balancing Technique",Counter(Y_train))
print("After applying Balancing Technique",Counter(Y_train_smote))
```

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
Lr = LogisticRegression()
```

```python
## checking cross validation score
scores = cross_val_score(Lr,X,Y,cv=5,scoring='accuracy')
print("Cross-Validation Scores:",scores)
print("Cross validation Mean Score:",scores.mean())
print("Std :",scores.std())
```

```python
Lr.fit(X_train_smote,Y_train_smote)
```

```python
pred = Lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score
```

```python
accuracy_score(Y_test,pred)
```

```python
f1_score(Y_test,pred,average='weighted')
```
